### 4) Symptoms

```python
#symptom distribution
covid_new.symptom.value_counts()
```

```python
#find the longest symptom
list = []
for i in covid_new.symptom:
    list.append(len(str(i).split(',')))
    
print(max(list))
```

```python
#define functions to seperate the symptoms(6 functions)
def divi_symptom1(x):
    return x.split(',')[0].strip()

def divi_symptom2(x):
    if len(x.split(','))>=2:
        return x.split(',')[1].strip()
    else:
        return np.nan
    
def divi_symptom3(x):
    if len(x.split(','))>=3:
        return x.split(',')[2].strip()
    else:
        return np.nan
    
def divi_symptom4(x):
    if len(x.split(','))>=4:
        return x.split(',')[3].strip()
    else:
        return np.nan
    
def divi_symptom5(x):
    if len(x.split(','))>=5:
        return x.split(',')[4].strip()
    else:
        return np.nan
    

def divi_symptom6(x):
    if len(x.split(','))>=6:
        return x.split(',')[5].strip()
    else:
        return np.nan
```

```python
#build new seperate symptom columns
covid_new.symptom = covid_new.symptom.astype(str)
covid_new['symptom1'] = covid_new['symptom'].apply(divi_symptom1)
covid_new['symptom2'] = covid_new['symptom'].apply(divi_symptom2)
covid_new['symptom3'] = covid_new['symptom'].apply(divi_symptom3)
covid_new['symptom4'] = covid_new['symptom'].apply(divi_symptom4)
covid_new['symptom5'] = covid_new['symptom'].apply(divi_symptom5)
covid_new['symptom6'] = covid_new['symptom'].apply(divi_symptom6)
```

# Modeling and training

## 1.Preprocessing

### 1) Encode the non-numerical values to numerical ones


```python
#set encoder
encoder = LabelEncoder()
covid_new.columns
```

```python
#creat function to encode
list = ['location', 'country', 'symptom1', 'symptom2', 'symptom3', 'symptom4', 'gender']
def encode_varible(list):
    for vari in list:
        covid_new[vari] = encoder.fit_transform(covid_new[vari])
encode_varible(list)  
```

```python
covid_new.head() 
covid_new.describe()
```

```python
covid_new.head()
```

### 2) Split data into feature and target

```python
#set death to target
covid_target = covid_new['death']
#covid_target = pd.DataFrame(covid_new['death'])

# set the rest to be features
covid_feature = covid_new.drop(columns='death')
covid_feature.head()
```

### 3) Feature scalling

```python
# from sklearn.preprocessing import StandardScaler to scale the data
#set scaler
scaler = StandardScaler()
#fit the data
covid_scaled = scaler.fit(covid_feature).transform(covid_feature)
#transfer np array to DataFrame
covid_scaled = pd.DataFrame(covid_scaled, columns = covid_feature.columns)
```

```python
covid_scaled.head()
```

### 4) Creating train test split

```python
# split the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(covid_scaled, covid_target, test_size=0.3, random_state=42)
```

```python
#check the shape of the splited data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 2. Select simple model and train

### 1) Logistic regression

```python
#select model
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
#train the model
model1.fit(X_train, y_train)
#predict using test set
predict1 = model1.predict(X_test)

#choose metric to evaluate the results
#Accuracy: (TP+TN)/ALL
lr_acc = metrics.accuracy_score(y_test, predict1)
print('The accuracy score of logistic regression is %10.4f' %lr_acc)
# Precision, recall, f1 and roc
#Sometimes Accuracy is not a good metric, especailly there are unbalanced sample
# Precision = TP/(TP+NP), Recall = TP/(TP+TN), f1 = 2 P*R/(P+R)
lr_prec = precision_score(y_test, predict1)
print('The precision score of logistic regression is %10.4f' %lr_prec)
lr_recall = recall_score(y_test, predict1)
print('The recall score of logistic regression is %10.4f' %lr_recall)
lr_f1 = f1_score(y_test, predict1)
print('The f1 score of logistic regression is %10.4f' %lr_f1)
lr_auc = roc_auc_score(y_test, predict1)
print('The roc_auc score of logistic regression is %10.4f' %lr_auc)
```

```python
#visualizztion the metrics
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.bar(['Accuracy', 'Precision', 'Recall', 'F1', 'Roc_auc'], [lr_acc, lr_prec, lr_recall, lr_f1, lr_auc], 
         color='green');
plt.title('Metrics comparision of logistic regression model')
```

```python
#print confusion matrix
from sklearn.metrics import plot_confusion_matrix
print('The confusion matrix of logestic regression algorithm:')
plot_confusion_matrix(model1,X_test, y_test);
plt.show()
```

### 2) SVM

```python
#select model
from sklearn.svm import SVC
model2 = SVC()
#train the model
model2.fit(X_train, y_train)
#predict using test set
predict2 = model2.predict(X_test)

#try different metrics
svm_acc = metrics.accuracy_score(y_test, predict2)
print('The accuracy score of svm model is %10.4f' %svm_acc)
svm_prec = metrics.precision_score(y_test, predict2)
print('The precision score of svm model is %10.4f' %svm_prec)
svm_recall = metrics.recall_score(y_test, predict2)
print('The recall score of svm model is %10.4f' %svm_recall)
svm_f1 = metrics.f1_score(y_test, predict2)
print('The f1 score of svm model is %10.4f' %svm_f1)
svm_auc = metrics.roc_auc_score(y_test, predict2)
print('The roc_auc score of svm model is %10.4f' %svm_auc)
```

```python
#plot the erros with different k
classification_error = [1-x for x in cv_value]

plt.plot(range(2,50),  classification_error)
plt.title('The misclassification rate by different k values')
plt.xlabel('K values of KNN')
plt.ylabel('Misclassification rate')

# find the best k
k = classification_error.index(min(classification_error))
print('The best k value with the min classification error is: %5.3f' %k)
```
